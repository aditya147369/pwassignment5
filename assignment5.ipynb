Q1, Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.

```python
class vehicle:
    def __init__(self,name_of_vehicle,max_speed,average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle
```

Q2. Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.

```python
class car(vehicle):
    def seating_capacity(self,capacity):
        return self.name_of_vehicle,capacity

car_obj = car("BMW",280,150)
car_obj.seating_capacity(7)
```

Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

```python
"Multiple inheritance is process in which a particular class can inherit multiple classes and it can use all the methods from those classes as well as it can use its own methods also"

class first:
    def first_method(self):
        print("this is 1st class method")

class second:
    def second_method(self):
        print("this is 2nd class method")

class third(first,second):
    def third_method(self):
        print("this is 3rd class method")

third_obj = third()
third_obj.first_method()
third_obj.second_method()
third_obj.third_method()
```

Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.

```python
"When we initialize a variable in a class constructor and set it as private for that we use self.__instance_name then we cannot set its value and display it directly for that we can use two fucntions set() and get()"

class set_get:

    def __init__(self,num):
        self.__num = num

    def set(self,set_num):
        self.__num=set_num

    def get(self):
        return self.__num

obj = set_get(5)
obj.set(10)
print(obj.get())
obj.set(20)
print(obj.get())
```

Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

```python
'''Method overriding means suppose we have two classes super class and sub class and the sub class is inherited from super class and both the class have a same function example display() so when we make a object of 
sub class and call the display() the sub class will hide the super class display() and execute the sub classes display() this method overriding ''' 

class parent:

    def display(self):
        print("hello")

class child(parent):

    def display(self):
        print("hello world")

obj = child()
obj.display()
```
